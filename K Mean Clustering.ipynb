```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Dataset

```python
df = pd.read_csv('adult(cleaned).csv')
```

### Metric Function

```python
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, precision_score, f1_score, ConfusionMatrixDisplay

def metrics(y_true, y_pred):
    print(f"accuracy : {accuracy_score(y_true, y_pred)}" )
    print(f"recall : {recall_score(y_true, y_pred)}" )
    print(f"precision : {precision_score(y_true, y_pred)}" )
    print(f"f1 : {f1_score(y_true, y_pred)}" )
    cm = confusion_matrix(y_true, y_pred)
    print(f"confusion matrix: {cm}")
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    plt.show()
```

### Train test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('income', axis=1)
y = df[['income']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y.values.ravel(), random_state=42, stratify=y)
```

### Importing the classifier

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans()
```

### Training with normalized dataset

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
metrics(y_test,y_pred)
```

### Training with non normalized dataset

```python
df_non_normalized = pd.read_csv('adult(cleaned non normalized).csv')
```

```python
x = df_non_normalized.drop('income', axis=1)
y = df_non_normalized[['income']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=42, stratify=y)
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```

### With Standardization

```python
from sklearn.preprocessing import StandardScaler

# Standardize the features
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled,y, random_state=42, stratify=y)
```

```python
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
accuracy_score(y_test, y_pred)
```
